# Telecomm Customer Churn

This data set provides information to help you predict what behavior will help you to retain customers. You can analyze all relevant customer data and develop focused customer retention programs.

It is requested a model to predict customer churn. Customers who left within the last month – the column is called Churn

Dataset: https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/ChurnData.csv

- Use Logistical Regression to build your model
- Features: 'tenure', 'age', 'address', 'income', 'ed', 'employ', 'equip'

```python
import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
from sklearn import preprocessing
%matplotlib inline 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import jaccard_score
from sklearn.metrics import log_loss
```

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/ChurnData.csv')
print(df.shape)
df.head()
```

```python
print(df.dtypes)
df.describe()
```

```python
df.corr(method ='pearson').sort_values(by = 'churn', ascending=False).head()
```

```python
# Features to array
X = np.asarray(df[['tenure', 'age', 'address', 'income', 'ed', 'employ', 'equip']])
X
```

```python
# Target data type to int - required to sklearn LR
#Target to array
df['churn'] = df['churn'].astype('int')
y = np.asarray(df['churn'])
y
```

```python
# Normalizing Features
X = preprocessing.StandardScaler().fit(X).transform(X)
```

```python
# Data Split 80/20
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
# Modeling and Training
lr = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)

# Prediction and Probability
lr_pred = lr.predict(X_test)
lr_proba = lr.predict_proba(X_test)


```

# Evaluation

```python
#Jaccard - binary - positive label = 0 (not churn)
jaccard = jaccard_score(y_test,lr_pred, pos_label=0)
jaccard
```

```python
# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test, lr_pred, labels=[1,0])
np.set_printoptions(precision=2)
print (classification_report(y_test, lr_pred))
cnf_matrix
```

```python
# Logloss
logloss = log_loss(y_test, lr_proba)
```

```python
print('Results: Jaccard= ', jaccard,'F1 = 0.72', 'LogLoss=', logloss, 'for regularization = 0.01 and solver liblinear' )
```
